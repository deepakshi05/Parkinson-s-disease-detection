```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# loading the data from csv file to a Pandas DataFrame
parkinsons_data = pd.read_csv('/content/parkinsons.data.csv')
```

```python
# printing the first 5 rows of the dataframe
parkinsons_data.head()
```

```python
# number of rows and columns in the dataframe
parkinsons_data.shape
```

```python
# getting more information about the dataset
parkinsons_data.info()
```

```python
# checking for missing values in each column
parkinsons_data.isnull().sum()
```

```python
# getting some statistical measures about the data
parkinsons_data.describe()
```

```python
# distribution of target Variable
parkinsons_data['status'].value_counts()
```

```python
# grouping the data based on the target variable
parkinsons_data.groupby('status').mean()
```

```python
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Histogram of a Feature


```python
plt.figure(figsize=(8, 6))
sns.histplot(data=parkinsons_data, x='MDVP:Fo(Hz)', hue='status', bins=20, kde=True)
plt.title("Distribution of 'MDVP:Fo(Hz)' Feature")
plt.xlabel("MDVP:Fo(Hz)")
plt.ylabel("Frequency")
plt.legend(['Healthy', 'PD Positive'])
plt.show()
```

# Bar Chart for Accuracy Scores

```python
accuracy_scores = [training_data_accuracy, test_data_accuracy]
datasets = ['Training Data', 'Test Data']

plt.figure(figsize=(8, 6))
sns.barplot(x=datasets, y=accuracy_scores)
plt.title("Model Accuracy on Training and Test Data")
plt.xlabel("Dataset")
plt.ylabel("Accuracy Score")
plt.ylim(0, 1.0)
plt.show()
```

Pair Plot for Feature Analysis

```python


sns.set(style="ticks")

# Define the features you want to include in the pair plot
features = [
    'MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
]

# Set the figure size for the pair plot
plt.figure(figsize=(10, 8))  # Adjust the width and height as needed

# Create the pair plot with adjusted layout and smaller graphs
pairplot = sns.pairplot(parkinsons_data[features + ['status']], hue='status', diag_kind='kde', height=1.5)  # Adjust the height value

# Adjust the position and spacing of the title
pairplot.fig.suptitle("Pair Plot of Selected Features\nby Disease Status", y=1.02)  # Use a newline character for spacing
pairplot.fig.subplots_adjust(top=0.9)  # Adjust the position of the suptitle
pairplot.fig.subplots_adjust(hspace=0.5)  # Adjust the vertical spacing between subplots

plt.tight_layout()
plt.show()
```

Box Plot for Feature Distribution

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='status', y='MDVP:Fo(Hz)', data=parkinsons_data)
plt.title("Box Plot of 'MDVP:Fo(Hz)' Feature by Disease Status")
plt.xlabel("Disease Status")
plt.ylabel("MDVP:Fo(Hz)")
plt.xticks([0, 1], ['Healthy', 'PD Positive'])
plt.show()
```

Count Plot for Disease Status

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=parkinsons_data, x='status')
plt.title("Count of Healthy and PD Positive Subjects")
plt.xlabel("Disease Status")
plt.ylabel("Count")
plt.xticks([0, 1], ['Healthy', 'PD Positive'])
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = parkinsons_data.corr(method='pearson', numeric_only=True)
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, yticklabels=True, xticklabels=True, annot_kws={"size": 8})
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.title("Correlation Heatmap")
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
